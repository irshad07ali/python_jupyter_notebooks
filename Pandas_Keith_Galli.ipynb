```python
bios = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
```

```python
bios.head()
```

```python
# filter height >210 and weight >170
```

```python
bios.info()
```

```python
bios.loc[bios["height_cm"] > 210]
```

```python
bios.loc[bios["height_cm"] > 220 ]
```

```python
bios.loc[bios["height_cm"] > 220 , ["name", "height_cm"] ]
```

```python
# similar results
bios.loc[bios["height_cm"] > 220][["name", "height_cm"]] # column list similar to array
```

```python
# similar results
bios.loc[(bios["height_cm"] > 210  ) & (bios["weight_kg"] > 100)][["name", "height_cm" , "weight_kg"]]
```

```python
# similar results
bios.loc[(bios["height_cm"] > 210  ) & (bios["born_country"] == "USA")][["name", "height_cm" , "weight_kg"]]
```

```python
# similar results
# bios.loc[(bios["height_cm"] > 210  ) and (bios["born_country"] == "USA")][["name", "height_cm" , "weight_kg"]]
```

```python
bios[bios["name"].str.contains("Keith")] # str.contains similar to " like clause" in sql
```

```python
bios[bios["name"].str.contains("keith" , case = False)] # str.contains similar to " like clause" in sql
```

```python
bios[bios["name"].str.contains("eith" , case = False)] # str.contains is similar to " like clause" in sql
```

```python
bios[bios["name"].str.contains("keith|patrik" , case = False)] # similar to " like clause" in sql keith or patrick
```

```python
# explore regex in pandas thru chatgpt - keith video
```

```python
bios[bios["name"].str.contains("keith|patrik" , case = False , regex = False)] # similar to " in clause" in sql keith or patrick
```
